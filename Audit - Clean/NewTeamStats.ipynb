```python
# use pandas lib to read csv file into a data frame
#https://github.com/nikbearbrown/INFO_6210/blob/master/Week_2/NBB_Data_cleaning_IMDB.ipynb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot 
df = pd.read_csv('OldTeamStats.csv')
df
```

```python
#drop columns that are empty and we don't need in this task
#https://www.geeksforgeeks.org/python-delete-rows-columns-from-dataframe-using-pandas-drop/
df.drop(["AssistsPercentage", "BlocksPercentage","DefensiveReboundsPercentage","DoubleDoubles","EffectiveFieldGoalsPercentage",
        "FantasyPoints","FantasyPointsDraftKings","FantasyPointsFanDuel","FantasyPointsFantasyDraft","FantasyPointsYahoo","GlobalTeamID",
        "IsClosed","OffensiveReboundsPercentage","OpponentPosition","OpponentStat","PlayerEfficiencyRating","PlusMinus","Possessions",
        "Season","SeasonType","Seconds","StatID","StealsPercentage","TeamID","TotalReboundsPercentage","TripleDoubles","TrueShootingAttempts",
        "TrueShootingPercentage","TurnOversPercentage","Updated","UsageRatePercentage"],axis = 1, inplace = True) 
df
```

```python
#view columns list and see if there is null value
#https://github.com/rajsankhe/python_workshop/blob/6c95fa3f441f42b0526f08f21e2b362671674bbe/Subtopics/9-Pandas.ipynb
df.columns.tolist()
df.isnull().sum()
```

```python
#Rename some columns's names to make format more consistent
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html
df.rename(columns={'BlockedShots': 'Blocks', 'Name': 'Team_Name', 'PersonalFouls': 'Fouls'}, inplace=True)
df[['Blocks', 'Team_Name','Fouls']].head()
```

```python
#Drop another unnecessary column
df.drop("Games", axis=1, inplace=True)
df.columns.tolist()
```

```python
#View Team's names to see if every team has the correct 3-letter abbreviation
df['Team']
```

```python
#edit teams'names which are not consistent
df.replace('GS', 'GSW', inplace = True)
df.replace('NO', 'NOP', inplace = True)
df.replace('NY', 'NYK', inplace = True)
df.replace('SA', 'SAS', inplace = True)
df['Team']
```

```python
#sort teams by wins record and give index
#https://stackoverflow.com/questions/33165734/update-index-after-sorting-data-frame
df=df.sort_values('Wins', ascending = False)
df = df.reset_index(drop=True)
df.index = df.index+1
df[['Team','Wins']].head()
```

```python
#change sequence of columns to make the sheet more resonable and readable
#https://erikrood.com/Python_References/change_order_dataframe_columns_final.html
df1=df[['Team_Name','Team','Wins','Losses','Minutes','Points',
        'FieldGoalsAttempted','FieldGoalsMade','FieldGoalsPercentage',
        'TwoPointersAttempted','TwoPointersMade','TwoPointersPercentage',
       'ThreePointersAttempted','ThreePointersMade','ThreePointersPercentage',
        'FreeThrowsAttempted', 'FreeThrowsMade', 'FreeThrowsPercentage',
       'Rebounds','OffensiveRebounds','DefensiveRebounds','Assists', 'Blocks','Steals','Turnovers','Fouls' ]]
df1
```

```python
#check if the range of numerical variables are reasonable
df1.describe()
```

```python
#Check if the FG% is reasonable
#https://github.com/nikbearbrown/INFO_6210/blob/master/Week_2/NBB_Data_Munging_Wrangling.ipynb
df1['FieldGoalsPercentage'].hist(bins=5)
```

```python
#Check if Points is reasonable
sns.boxplot(x=df1["Points"])
```

```python
#save df1 to a csv file
#https://github.com/nikbearbrown/INFO_6210/blob/master/Week_2/NBB_Data_cleaning_IMDB.ipynb
df1.to_csv('NewTeamStats.csv')
```

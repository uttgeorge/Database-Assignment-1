```python
import pandas as pd
df = pd.read_csv('OldPlayerDetails.csv')
```

```python
df.shape
```

```python
df.columns.tolist()
df.isnull().sum()
```

```python
df.rename(columns={'YahooName': 'PlayerName', 'Jersey': 'JerseyNo','Experience':'ExperienceYear'}, inplace=True)
df[['PlayerName','JerseyNo','ExperienceYear']].head()
```

```python
df1=df[['PlayerName','Team','BirthDate','BirthCity','College','Height','Weight','PhotoUrl'
      ,'Position','JerseyNo','ExperienceYear']]
df1
```

```python
#https://stackoverflow.com/questions/13682044/remove-unwanted-parts-from-strings-in-a-column
df1['BirthDate']=df1['BirthDate'].map(lambda x: str(x)[0:10])
```

```python
#https://www.geeksforgeeks.org/python-pandas-dataframe-fillna-to-replace-null-values-in-dataframe/
df1["College"].fillna("No College", inplace = True)
df1["College"]
```

```python
df1["BirthCity"].fillna("No record", inplace = True)
df1['BirthCity']
```

```python
df1["BirthDate"].fillna("No record", inplace = True)
df1['BirthDate']
```

```python
df['Team'].unique()
```

```python
df1.replace('GS', 'GSW', inplace = True)
df1.replace('SA', 'SAS', inplace = True)
df1.replace('NY', 'NYK', inplace = True)
df1.replace('NO', 'NOP', inplace = True)
df1
```

```python
df1.to_csv('NewPlayerDetails.csv', index=False)
```

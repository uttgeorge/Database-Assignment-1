```python
#use pandas lin to read csv file in a data frame
#https://github.com/nikbearbrown/INFO_6210/blob/master/Week_2/NBB_Data_cleaning_IMDB.ipynb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot 
df = pd.read_csv('OldPlayerDetails.csv')
```

```python
#see size of data frame
df.shape
```

```python
#see how many null values in columns
#https://github.com/rajsankhe/python_workshop/blob/6c95fa3f441f42b0526f08f21e2b362671674bbe/Subtopics/9-Pandas.ipynb
df.columns.tolist()
df.isnull().sum()
```

```python
#Rename columns' names that are misleading
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html
df.rename(columns={'YahooName': 'PlayerName', 'Jersey': 'JerseyNo','Experience':'ExperienceYear'}, inplace=True)
df[['PlayerName','JerseyNo','ExperienceYear']].head()
```

```python
#change sequence of columns to make the sheet more resonable and readable
#https://erikrood.com/Python_References/change_order_dataframe_columns_final.html
df1=df[['PlayerName','Team','BirthDate','BirthCity','College','Height','Weight','PhotoUrl'
      ,'Position','JerseyNo','ExperienceYear']]
df1
```

```python
#Revise the format of birth date, keeping the first 10 str
#https://stackoverflow.com/questions/13682044/remove-unwanted-parts-from-strings-in-a-column
df1['BirthDate']=df1['BirthDate'].map(lambda x: str(x)[0:10])
df1['BirthDate']
```

```python
#Fill 'No College' in the missing value positions of variable College
#https://www.geeksforgeeks.org/python-pandas-dataframe-fillna-to-replace-null-values-in-dataframe/
df1["College"].fillna("No College", inplace = True)
df1["College"]
```

```python
#fill 'No record' in the missing value position of variable BirthCity
df1["BirthCity"].fillna("No record", inplace = True)
df1['BirthCity']
```

```python
#fill 'No record' in the missing value position of variable BirthDate
df1["BirthDate"].fillna("No record", inplace = True)
df1['BirthDate']
```

```python
#View unique teams'names to see if format is consistent
#https://chrisalbon.com/python/data_wrangling/pandas_list_unique_values_in_column/
df['Team'].unique()
```

```python
#Revise the teams's names that are not consistent
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html
df1.replace('GS', 'GSW', inplace = True)
df1.replace('SA', 'SAS', inplace = True)
df1.replace('NY', 'NYK', inplace = True)
df1.replace('NO', 'NOP', inplace = True)
df1
```
